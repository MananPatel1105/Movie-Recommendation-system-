## 1) <b> Demographic Filtering </b>: - <br> 
##### They offer generalized recommendations to every user, based on movie popularity and/or genre. This approach is considered to be too simple. The basic idea behind this system is that movies that are more popular will have a higher probability of being liked by the average audience.

__________________________________________________________________________________________________________________________________________________________________________________

##### For this we will use <b> Weighted Rating </b> Formula which is given by: - 
<img src="image.png"> <br><br> ![Alt text](image-1.png)

```python
m = data['vote_count'].quantile(0.90) #This means that movie votes must be more than 90% of the other movie votes.
C = data['vote_average'].mean()
```

```python
#Filtering Movies according to our threshold
filtered_movies = data.copy()
filtered_movies = filtered_movies[filtered_movies['vote_count'] > m]
print(f'Original Dataset: - {data.shape}\n\nFiltered Dataset: - {filtered_movies.shape}')
```

```python
#Function for calculating Weighteed Average
def weighted_average(x, m = m, C = C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
filtered_movies['Score'] = filtered_movies.apply(weighted_average, axis=1)
```

```python
filtered_movies = filtered_movies[['title', 'Score']].sort_values(by=['Score'], ascending=False,).head(10)
filtered_movies = filtered_movies.reset_index()
filtered_movies.drop(columns=['index'], inplace=True)
filtered_movies
```

## 2) <b> Content-Based Filtering </b>: - <br> 
##### They suggest similar items based on a particular item. The general idea behind these recommender systems is that if a person liked a particular item, he or she will also like an item that is similar to it.

_________________________________________________________________________________________________________________________________________________________________________________

```python
#Calculating the TF-IDF Score
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(data['Summary']).toarray()

tfidf_matrix.shape
```

```python
tfidf_matrix
```

```python
#Calculating the Similarity Score (Cosine Similarity)
from sklearn.metrics.pairwise import cosine_similarity

cosine_matrix = cosine_similarity(tfidf_matrix)
```

```python
cosine_matrix.shape
```

```python
#This is basically the cosine score of each movie with the other movies
cosine_matrix
```

```python
#Creating Function for Recommendation
indices = np.arange(0,4799,1)
def recommend(movie):
    index_movie = data[data['title'] == movie].index[0]    #Index of the particular movie

    indices_and_movies = [(i, j) for i,j in zip(indices, cosine_matrix[index_movie])]   # Movie with high similarity score with there indices

    indices_and_movies = sorted(indices_and_movies, reverse=True, key=lambda x:x[1])[1:6]    #Sorting them with there score and not taking the first one as it is the movie itself.

    movies = [data.loc[i[0], 'title'] for i in indices_and_movies]

    return pd.DataFrame({"Movies":movies, 'Scores':[i[1] for i in indices_and_movies]}, index=[1,2,3,4,5])
```

Final Recommendataions

```python
#Final Recommendations
recommend("Harry Potter and the Half-Blood Prince")
```

## 3) <b> Collaborative Filtering </b>: - <br> 
#####  This system matches persons with similar interests and provides recommendations based on this matching. Collaborative filters do not require item metadata like its content-based counterparts.

______________________________________________________________________________________________________________________________________________________________________________

#### <b>User-Based collaborative filtering</b> :- User-Based Collaborative Filtering is a recommendation system technique that suggests items to a target user based on the preferences and behaviors of other users who are similar to that target user.<br>

```python
ratings = pd.read_csv("E:/Extra Codes/Data Sciene/Datasets/Movie Recommendation System/Collaborative Filtering Dataset/ratings.csv")
movies = pd.read_csv("E:/Extra Codes/Data Sciene/Datasets/Movie Recommendation System/Collaborative Filtering Dataset/movies.csv")
```

```python
ratings.head()
```

```python
movies.head()
```

```python
data2 = pd.merge(ratings, movies, on='movieId', how='inner')
data2.drop(columns=['genres', 'timestamp'], inplace=True)
```

```python
data2.head()
```

#### <b> User-Based Collaborative filtering </b>

```python
#Creating the Sparse Matrix
matrix = data2.pivot_table(index='userId', columns='title', values='rating')
```

```python
matrix
```

```python
#Normalizing the Dataset
matrix = matrix.subtract(matrix.mean(axis=1), axis='rows')
matrix
```

```python
#if you want to use cosine Similarity

#Calculating the Similarity Matrix (Cosine Similarity)

'''user_similarity_matrix = cosine_similarity(matrix.fillna(0))
user_similarity_matrix = pd.DataFrame(user_similarity_matrix)
user_similarity_matrix'''
```

```python
#Calculating the Similarity Matrix (Person Similarity) between User to user
user_similarity_matrix = matrix.T.corr()
user_similarity_matrix
```

```python
#Picking a user and drop that user from matrix
picked_userid = 1
user_similarity_matrix.drop(index=picked_userid, inplace=True)

user_similarity_matrix.head()
```

```python
#number of similar users
n = 10

#Setting a threshold to select the similar users above this threshold (picking top 10)
user_similarity_threshold = 0.3

similar_users = user_similarity_matrix[user_similarity_matrix[picked_userid]>user_similarity_threshold][picked_userid].sort_values(ascending=False)[:n]

similar_users
```

```python
#Removing the Already Watched movies by user
picked_user_watched = matrix[matrix.index == picked_userid].dropna(axis=1, how = 'all')
picked_user_watched
```

```python
#Now removing the movies which none of the similar users have watched (not by the picked user and not by his similar users)
similar_users_movies = matrix[matrix.index.isin(similar_users.index)].dropna(axis=1, how='all')
similar_users_movies
```

```python
#Now dropping the Movies which the picked user have already watched from the similar_users_movies
similar_users_movies.drop(picked_user_watched.columns, axis=1, inplace=True, errors='ignore')
similar_users_movies
```

```python
#Now applying the weighted Average formula to calculate the Score


''' 
1) Picking the movie 
2) Now for that particular movie calculating the average score by multiplying the user's rating with the similarity score of that user with picked user
3) Finally averaging the rating.    
4) Creating a Dataframe with Movie name and Movie_Score
'''

item_score = {}
for i in similar_users_movies:
    movie_rating = similar_users_movies[i]
    total = 0
    count = 0

    for j in similar_users.index:
        if pd.isna(movie_rating[j]) == False:
            score = similar_users[j] * movie_rating[j]

            total += score
            count += 1

    item_score[i] = total/count

item_score = pd.DataFrame(item_score.items(), columns=['Movie', 'Movie_Score'])

item_score = item_score.sort_values(by=['Movie_Score'], ascending=False)

print(f'Movies Recommended for User_ID: - {picked_userid}')
item_score.head(10).reset_index().drop(columns=['index'])
```
